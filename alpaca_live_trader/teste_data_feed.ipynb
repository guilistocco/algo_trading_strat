```python
import sys
from pathlib import Path

# Garante que o diretório /home/ubuntu/algo_trading_strat esteja no PYTHONPATH
ROOT = Path.cwd().parent.resolve()
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from datetime import datetime, timedelta
import pandas as pd
from alpaca.data.requests import CryptoBarsRequest
from alpaca.data.historical import CryptoHistoricalDataClient

from alpaca.data.timeframe import TimeFrame

from alpaca_oms.alpaca_client import get_data_client
from config_etf import *
```

```python
end = datetime.utcnow()
limit = 100
start = end - timedelta(minutes=limit + 5)
# client = get_data_client()
client = CryptoHistoricalDataClient(APCA_API_KEY_ID, APCA_API_SECRET_KEY)


request = CryptoBarsRequest(
    symbol_or_symbols=SYMBOL,
    timeframe=TimeFrame.Minute,
    start=start,
    end=end,
    adjustment="all",
    feed="iex"
)
bars = client.get_crypto_bars(request).df
df = bars[SYMBOL] if isinstance(bars.columns, pd.MultiIndex) else bars
df
```
